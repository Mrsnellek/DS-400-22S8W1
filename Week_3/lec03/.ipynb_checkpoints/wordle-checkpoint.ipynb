```python
L5v = []
vowels = {"a", "e", "i", "o", "u", "y", "E", "A", "I", "O", "U"}

for word in l5:
    if sum(char in vowels for char in word)<2:
        result = "".join(dict.fromkeys(word))
        if len(result) == 5:
            L5v.append(result)
```
